##視覺化考拉茲猜想
我的prompt是"生成一個視覺化考拉茲猜想的python程式"。測試的AI一樣是GPT-4o和Grok3

可以從以下的生成結果看到在同樣的問題之下Grok3在一開始從尺度的設計、例子的數量以及美觀上都優於GPT-4o許多。

以下是完整程式碼

###GPT-4o

```python
import matplotlib.pyplot as plt

def collatz(n):
    """計算考拉茲數列，返回數列"""
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

def plot_collatz(n):
    """繪製考拉茲數列"""
    sequence = collatz(n)
    plt.figure(figsize=(10, 5))
    plt.plot(sequence, marker='o', linestyle='-', color='b')
    plt.xlabel("步數")
    plt.ylabel("數值")
    plt.title(f"考拉茲猜想：n = {n}")
    plt.grid()
    plt.show()

# 測試數字，例如 27
plot_collatz(27)
```

###Grok3

```python
import matplotlib.pyplot as plt
import numpy as np

def collatz(n):
    """生成單個數字的考拉茲序列"""
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

def plot_collatz_sequences(start_numbers):
    """繪製多個起始數字的考拉茲序列"""
    plt.figure(figsize=(12, 6))

    for start in start_numbers:
        sequence = collatz(start)
        plt.plot(sequence, label=f'Start: {start}', marker='o', alpha=0.6)

    plt.title('Collatz Conjecture Visualization')
    plt.xlabel('Step')
    plt.ylabel('Value')
    plt.yscale('log')  # 使用對數尺度，因為數值可能變化很大
    plt.grid(True, which="both", ls="-", alpha=0.2)
    plt.legend()
    plt.show()

# 測試一些起始數字
start_numbers = [7, 13, 19, 27, 42]

# 執行視覺化
plot_collatz_sequences(start_numbers)

# 額外：打印每個序列的長度和最大值
for start in start_numbers:
    seq = collatz(start)
    print(f"Starting number: {start}")
    print(f"Sequence length: {len(seq)}")
    print(f"Maximum value: {max(seq)}")
    print(f"First few terms: {seq[:min(5, len(seq))]}...")
    print()
```

```python
import matplotlib.pyplot as plt

def collatz(n):
    """計算考拉茲數列，返回數列"""
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

def plot_collatz(n):
    """繪製考拉茲數列"""
    sequence = collatz(n)
    plt.figure(figsize=(10, 5))
    plt.plot(sequence, marker='o', linestyle='-', color='b')
    plt.xlabel("步數")
    plt.ylabel("數值")
    plt.title(f"考拉茲猜想：n = {n}")
    plt.grid()
    plt.show()

# 讓使用者輸入數字
def main():
    try:
        n = int(input("請輸入一個正整數："))
        if n <= 0:
            print("請輸入大於 0 的正整數！")
        else:
            plot_collatz(n)
    except ValueError:
        print("請輸入有效的整數！")

# 開始互動
main()
```

###後續Grok3

```python
import matplotlib.pyplot as plt
import numpy as np

def collatz(n):
    """生成單個數字的考拉茲序列"""
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence

def get_valid_input(prompt, min_value=1):
    """確保使用者輸入有效正整數"""
    while True:
        try:
            value = int(input(prompt))
            if value < min_value:
                print(f"請輸入大於或等於 {min_value} 的正整數！")
                continue
            return value
        except ValueError:
            print("請輸入一個有效的整數！")

def interactive_collatz():
    """互動式考拉茲猜想視覺化"""
    print("歡迎體驗考拉茲猜想視覺化！")

    # 詢問使用者要測試多少個數字
    num_sequences = get_valid_input("請輸入要測試的起始數字數量（例如 1 到 10）：")

    # 收集起始數字
    start_numbers = []
    for i in range(num_sequences):
        prompt = f"請輸入第 {i+1} 個起始數字："
        start_num = get_valid_input(prompt)
        start_numbers.append(start_num)

    # 繪製圖表
    plt.figure(figsize=(12, 6))
    for start in start_numbers:
        sequence = collatz(start)
        plt.plot(sequence, label=f'Start: {start}', marker='o', alpha=0.6)

    plt.title('Collatz Conjecture Visualization')
    plt.xlabel('Step')
    plt.ylabel('Value')
    plt.yscale('log')  # 使用對數尺度
    plt.grid(True, which="both", ls="-", alpha=0.2)
    plt.legend()
    plt.show()

    # 顯示序列資訊
    print("\n序列統計資訊：")
    for start in start_numbers:
        seq = collatz(start)
        print(f"起始數字: {start}")
        print(f"序列長度: {len(seq)}")
        print(f"最大值: {max(seq)}")
        print(f"前幾項: {seq[:min(5, len(seq))]}...")
        print()

# 運行互動程式
if __name__ == "__main__":
    interactive_collatz()
```
