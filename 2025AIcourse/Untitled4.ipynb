```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定義 x 軸範圍，從 -100 到 100，共 100 個點
x = np.linspace(-100, 100, 100)
y = np.tan(x)
# 繪製圖形，使用紅色並加粗線條
plt.plot(x, y, color='red', linewidth=2.5)
# 顯示網格線
plt.grid(True)
plt.show()
```
